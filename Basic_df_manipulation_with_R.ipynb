## Dataframe manipulation with R

```r
#Creating a dataframe with only name and bodyweight, arranged by bodyweight in highest to lowest order

Q1 <- msleep |>
  select(name, bodywt) |>
  arrange(desc(bodywt)) |>
  as.data.frame()

Q1
```

```r
#Finding the animal with the highest bodyweight

Q3 <- msleep |>
  select(name, bodywt) |>
  arrange(desc(bodywt)) |>
  select(name) |>
  slice_head(n=1) |>
  as.data.frame()
  
Q3
```

```r
#Finding the animal with the lowest brainweight

Q4 <- msleep |>
  select(name, brainwt) |>
  arrange(brainwt) |>
  select(name) |>
  slice_head(n=1) |>
  as.data.frame()




### VIEW OUTPUT ###
Q4
```

```r
#Creating a dataframe and sorting by genus while also filtering for only primates

Q5 <- msleep |>
  filter(order == "Primates") |>
  select(genus, name, order) |>
  arrange(genus) |>
  as.data.frame()

Q5
```

```r
#Filtering a dataframe for only carnivores with the conservation status of LC and calculating the variance in total sleep

Q6 <- msleep |>
  filter(vore == "carni" & conservation == "lc") |>
  group_by(vore) |>
  summarise(var = round(var(sleep_total),2)) |>
  select(var) |>
  as.data.frame()

Q6
```

```r
#Filtering for rodent with the highest ratio of sleep to REM sleep

Q7 <- msleep |>
  filter(order == "Rodentia") |>
  mutate(sleep_ratio = sleep_total/sleep_rem) |>
  slice_max(sleep_ratio, n=1) |>
  select(name) |>
  as.data.frame()

Q7
```

```r
#Finding how many primates have a bodyweight to brain weight ratio greater than 100

Q8 <- msleep |>
  filter(order == "Primates") |>
  mutate(brain_body_ratio = bodywt/brainwt) |>
  filter(brain_body_ratio > 100)|>
  count() |>
  as.data.frame()

Q8
```

```r
#Removing missing values, and finding the mean sleep and variance of sleep group

Q9 <- msleep |>
  filter(conservation != "NA") |>
  group_by(conservation) |>
  summarise(mean_sleep = round(mean(sleep_total, na.rm=TRUE),2),
            var_sleep = round(var(sleep_total, na.rm=TRUE),2)) |>
  as.data.frame()

Q9
```
